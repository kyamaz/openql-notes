## 4.5 計算複雑性理論と量子アルゴリズム

### 4.5.1 問題の難しさを考える

問題の難しさ　＝　あるアルゴリズムを実行するときに必要な計算リソースの計算量  
　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　・時間計算量（計算ステップ数）  
　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　　・空間計算量（メモリ量）  

計算機　・・・　チューリングマシンなどの計算モデルを仮定  
問題サイズなどを変化させたときの、必要計算リソース量の漸近的挙動で評価  


参考：[量子計算理論 量子コンピュータの原理, 森前/森北出版](https://www.amazon.co.jp/dp/4627854013)


**判定問題**

YESかNOで答えられる問題


**「P」の定義**

ある言語 L が P に入るとは、多項式時間決定的チューリングマシンが存在して、

1. $ x \in L $ のとき、受理

2. $ x \notin L $ のとき、拒否

を満たすことである。


**「BPP」の定義**

ある言語 L が BPP に入るとは、多項式時間確率的チューリングマシンが存在して、

1. $ x \in L $ のとき、受理確率 $ \ge 2/3 $

2. $ x \notin L $ のとき、受理確率 $ \le 1/3 $

を満たすことである。


確率的古典計算を多項式時間行うのは、我々が「可能」なことなので、ある意味、
<div style="color: red;">「簡単に解ける問題のクラス＝BPP」</div>


**「NP」の定義**

ある言語 L が NP に入るとは、多項式時間確率的チューリングマシンが存在して、

1. $ x \in L $ のとき、受理確率 $ \gt 0 $

2. $ x \notin L $ のとき、受理確率 $ = 0 $

を満たすことである。


### 4.5.2 量子コンピュータ版の複雑性クラス

BQP＝BPPの定義において、多項式時間確率的チューリングマシンを多項式時間量子計算機に変えたもの
つまり、

**「BQP」の定義**

ある言語 L が BQP に入るとは、多項式時間量子計算機が存在して、

1. $ x \in L $ のとき、受理確率 $ \ge 2/3 $

2. $ x \notin L $ のとき、受理確率 $ \le 1/3 $

を満たすことである。

<div style="color: red;">「量子計算で簡単に解ける問題のクラス＝BQP」</div>


**「PSPACE」とは、多項式サイズのメモリで解ける問題のクラス**



※確率古典計算機は量子計算の特別な場合なので、 $BPP \subseteq BQP $

※経路積分を使えば、 $BQP \subseteq PSPACE $ は示せる。

↓

$P \subseteq BPP \subseteq BQP \subseteq PSPACE $

<br/>

参照：[Strengths and Weaknesses of Quantum Computing, Charles H. Bennett, 1997](https://arxiv.org/pdf/quant-ph/9701001.pdf)

<br/><br/><br/>

<div style="color: red;">「量子計算機は古典計算機より速いか？」という問いは、</div>      
<div style="color: red;">判定問題の意味では「$BPP \neq BQP$か？」ということ・・・未解決</div>

**量子対話型証明系**

対話型証明系　:=　計算能力に制限のある検証者（verifier）が、計算能力に制限のない証明者（prover）とメッセージをやりとりすることにより、検証者の能力を超えた問題を解く系


**「MA」の定義**

言語 L が $MA $ に入るとは、ある多項式時間確率的チューリングマシンと多項式 $ w $ が存在して、

1. $ x \in L $ のとき、ある $ y \in \{ 0, 1 \} ^w $ が存在して、 $ (x, y) $ を入力として行う計算の受理確率は $ \ge 2/3 $

2. $ x \notin L $  のとき、どんな  $ y \in \{ 0, 1 \} ^w $  に対しても、受理確率は $ \le 1/3$

を満たすことである。


**「QMA」の定義**

言語 L が $QMA_w(c,s) $ に入るとは、ある多項式時間一様生成可能量子回路族 $ \{V_x\}_x $ と多項式 $ w, m$ が存在して、

1. $ x \in L $ のとき、ある $w$量子ビット状態 $\lvert \psi \rangle $ が存在して、受理確率

$$ (\langle \psi \rvert \otimes \langle 0^m \rvert ) V _x ^{\dagger} ( \lvert 0 \rangle \langle 0 \lvert  \otimes I ^{\otimes w + m -1} ) V_x ( \lvert \psi \rangle \otimes \lvert 0 \rangle ^ { \otimes m} )$$

が $ \ge c $

2. $ x \notin L $ のとき、どんな $w$量子ビット状態 $\lvert \psi \rangle $ に対しても、上記の受理確率は $ \le s$

を満たすことである。



* QMA-complete （QMA完全）

2-SAT1問題の例題：

$ f(x_1,x_2,x_3) = (\lnot x_1 \lor x_2) \land ( x_1 \lor x_3) \land (\lnot x_2 \lor x_3) \land (\lnot x_1 \lor \lnot  x_3) $

Hamiltonian への対応づけ：

$ H_s = H_1 + H_2 + H_3 + H_4 \\
= (\lvert 100 \rangle \langle 100 \rvert + \lvert 101 \rangle \langle 101 \rvert) + (\lvert 000 \rangle \langle 000 \rvert + \lvert 010 \rangle \langle 010 \rvert ) \\
+ ( \lvert 011 \rangle \langle 011 \rvert + \lvert 111 \rangle \langle 111 \rvert ) + ( \lvert 101 \rangle \langle 101 \rvert + \lvert 111 \rangle \langle 111 \rvert )$
